```python
from pyspark.sql import SparkSession
import dbldatagen as dg
```

```python
# Change "parquet" to "csv" and add the header option

dfSource = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("mode", "PERMISSIVE") \
    .load("../data/processed/dataset2_processed.csv")

# If it still fails, check the column names
dfSource.printSchema()

analyzer = dg.DataAnalyzer(sparkSession=spark, df=dfSource)

generatedCode = analyzer.scriptDataGeneratorFromData()
print(generatedCode)
```

```python
dfSource.printSchema()
```

```python
from pyspark.sql.functions import col

# 1. Filter out any row where a numeric column contains the literal string "double"
# We'll check the first column as a test
df_cleaned = dfSource.filter(col("patient_cc").isNotNull())

# 2. Specifically exclude rows that might be string headers accidentally caught in the data
df_cleaned = df_cleaned.filter(col("patient_cc").cast("string") != "double")

# 3. Cache it to break the lineage and ensure Spark processes the filter NOW
df_cleaned.cache()

# 4. Now run the analyzer on the cleaned data
analyzer = dg.DataAnalyzer(sparkSession=spark, df=df_cleaned)

generatedCode = analyzer.scriptDataGeneratorFromData()
print(generatedCode)
```

```python
# This will show you rows that cannot be converted to a number properly
dfSource.select("patient_cc").distinct().show()
```

```python
import dbldatagen as dg

# 1. Use the schema you already have
# (Since we know the data is clean based on your 'distinct' check)
source_schema = dfSource.schema

# 2. Use DataAnalyzer's schema-based method instead of data-based
# This avoids the casting/scanning that is causing the NumberFormatException
generatedCode = dg.DataAnalyzer.scriptDataGeneratorFromSchema(
    source_schema, 
    tableName="generated_patients"
)

print(generatedCode)
```
