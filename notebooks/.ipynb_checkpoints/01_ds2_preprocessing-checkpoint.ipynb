#### Preprocessing steps:
1. DONE: check distirbution, data card was missing in kaggle - to check balance
2. DONE: there are no null values, just deal with encoding
3. target is diagnosis: 1 (healthy), 2 (benign), 3 (pancreatic cancer)
4. DONE: drop: sample id, stage(?) - check if the values of stage and diagnosis match - they do so drop to prevent leakage
5. DONE: encode: sex, patient_cc, sample_o

```python
import pandas as pd
path = '../data/raw/pancreatic_600_risk123.csv'
df = pd.read_csv(path)

df
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
#preventing data leakage -> drop stage
print(pd.crosstab(df['diagnosis'], df['stage']))
```

The crosstab shows a perfect 1-to-1 mapping between stage and diagnosis. They encode exactly the same information, just with different granularity (stage splits diagnosis=3 into subcategories... wait, actually here diagnosis is stage — they're the same variable).

```python
df = df.drop(columns=['stage', 'sample_id'])
```

```python
df.info()
```

```python
#encode patient_cc
print(df['patient_cc'].value_counts())
print(f"\nUnique values: {df['patient_cc'].nunique()}")
print(df['patient_cc'].unique())
```

```python
df['patient_cc'] = df['patient_cc'].map({'Cohort1': 0, 'Cohort2': 1, 'Cohort3': 2})
df.info()
```

```python
#encode sex
print(df['sex'].unique())
df['sex'] = df['sex'].map({'M': 0, 'F': 1})
df.info()
```

```python
#encode sample_o
print(df['sample_o'].value_counts())
print(f"\nUnique values: {df['sample_o'].nunique()}")
print(df['sample_o'].unique())
```

```python
df['sample_o'] = df['sample_o'].map({'BPTB': 0, 'CPTB': 1, 'UPTB': 2})
df.info()
```

```python
#convert all to float
df = df.astype('float64')
df.info()
```

```python
#save to processed
df.to_csv('../data/processed/dataset2_processed.csv', index=False)
```

```python
df.skew().sort_values(ascending=False).round(2)
#nothing too skewed
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns

#there are no extreme outliers as we can see from the plots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 12))
axes = axes.flatten()

cols = [c for c in df.columns if c != 'diagnosis']
for i, col in enumerate(cols):
    sns.boxplot(x=df['diagnosis'], y=df[col], ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```
